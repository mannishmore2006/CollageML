```python
import numpy as py
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Heart.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['MaxHR'], c='blue', alpha=0.5)
plt.title('Scatter Plot: Age vs MaxHR')
plt.xlabel('Age')
plt.ylabel('MaxHR')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df['AHD'].value_counts().plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Bar Plot: Distribution of Heart Disease')
plt.xlabel('Heart Disease (Yes/No)')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df.boxplot(column='RestBP', by='AHD', patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.title('Box Plot: RestBP Distribution by Heart Disease')
plt.suptitle('')
plt.xlabel('Heart Disease (Yes/No)')
plt.ylabel('Resting Blood Pressure')
plt.show()
```

```python
sex_counts = df['Sex'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(sex_counts, labels=['Male', 'Female'], autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('Pie Chart: Sex Distribution')
plt.show()
```

```python
df.groupby('ChestPain').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(x='Age', y='MaxHR', hue='AHD', data=df, marker='o', palette='coolwarm')
plt.title('Age vs Max Heart Rate (MaxHR) by Heart Disease (AHD)')
plt.xlabel('Age')
plt.ylabel('Max Heart Rate')
plt.show()
```
