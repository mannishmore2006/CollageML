```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Generate synthetic data
np.random.seed(42)
num_samples = 200
data = {
    'Age': np.random.randint(18, 65, num_samples),
    'PurchaseFrequency': np.random.randint(1, 20, num_samples),
    'AnnualSpending': np.random.uniform(100, 2000, num_samples),
    'SpecialOfferResponse': np.random.choice([0, 1], num_samples)
}
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
# Features and target
X = df.drop(columns='SpecialOfferResponse')
y = df['SpecialOfferResponse']
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Apply Random Forest Classifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
# Predictions
y_pred = model.predict(X_test)
```

```python
# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, cmap='Blues', fmt='d')
plt.title("Confusion Matrix")
plt.show()
```

```python
# Evaluation Metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
```

```python

print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
```
